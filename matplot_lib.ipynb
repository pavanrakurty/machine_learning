```python
#import matplotlib library

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)

print(x)
print(y)
print(z)
```

```python
# sine wave
plt.plot(x, y)
plt.show()
```

```python
#cosine wave
plt.plot(x, z)
plt.show()
```

```python
#adding labels to X and Y axis
plt.plot(x, y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#parabola

x = np.linspace(-10,10,20)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
#plot with diff color
plt.plot(x, y, 'r+')
plt.show()
```

```python
#plot with diff color
plt.plot(x, y, 'rx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```
